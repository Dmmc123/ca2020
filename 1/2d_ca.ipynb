```python
import random
from PIL import Image, ImageDraw
```

```python
n = 128
size = 512
k = size // n

aut = [[0] * n for i in range(n)]
for i in range(n):
    for j in range(n):
        if (random.randint(0, 1) == 0): aut[i][j] = 1
```

```python
img = Image.new('RGB', (size, size), color = 'white')

draw = ImageDraw.Draw(img)
for i in range(n):
    for j in range(n):
        if (aut[i][j] == 1): draw.rectangle([(k*j, k*i), (k*(j+1)-1, k*(i+1)-1)], fill = "black") 
del draw

display(img)
```

```python
while (input() != '0'):
    temp_aut = [[0] * n for i in range(n)]
    
    for i in range(n):
        for j in range(n):
            # counting neighbors
            nbr = 0
            if (aut[(i - 1) % n][(j - 1) % n] == 1): nbr += 1
            if (aut[(i - 1) % n][j] == 1): nbr += 1
            if (aut[(i - 1) % n][(j + 1) % n] == 1): nbr += 1
            if (aut[i][(j - 1) % n] == 1): nbr += 1
            if (aut[i][(j + 1) % n] == 1): nbr += 1
            if (aut[(i + 1) % n][(j - 1) % n] == 1): nbr += 1
            if (aut[(i + 1) % n][j] == 1): nbr += 1
            if (aut[(i + 1) % n][(j + 1) % n] == 1): nbr += 1
                
            if (aut[i][j] == 0):
                if (nbr == 3): temp_aut[i][j] = 1
            else:
                if (nbr == 2 or nbr == 3): temp_aut[i][j] = 1
                    
    aut = temp_aut.copy()
    
    img = Image.new('RGB', (size, size), color = 'white')
    
    draw = ImageDraw.Draw(img)
    for i in range(n):
        for j in range(n):
            if (aut[i][j] == 1): draw.rectangle([(k*j, k*i), (k*(j+1)-1, k*(i+1)-1)], fill = "black") 
    del draw
    
    display(img)
```
