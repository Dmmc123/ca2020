```python
import random
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
```

```python
n = 128
size = 512
k = size // n

Sus = 0
Inf = 1
Rec = 2

Inf_period = 14

Sus_color = "blue"
Inf_color = "red"
Rec_color = "green"

aut = [[Sus] * n for i in range(n)]
inf_days = [[0] * n for i in range(n)]

aut[n//2][n//2] = Inf
inf_days[n//2][n//2] = Inf_period

# for i in range(n):
#     for j in range(n):
#         if (random.randint(0, 1) == 0): aut[i][j] = 1
```

```python
img = Image.new('RGB', (size, size), color = 'white')

draw = ImageDraw.Draw(img)
for i in range(n):
    for j in range(n):
        if (aut[i][j] == Sus): draw.rectangle([(k*j, k*i), (k*(j+1)-1, k*(i+1)-1)], fill = Sus_color) 
        elif (aut[i][j] == Inf): draw.rectangle([(k*j, k*i), (k*(j+1)-1, k*(i+1)-1)], fill = Inf_color)
        elif (aut[i][j] == Rec): draw.rectangle([(k*j, k*i), (k*(j+1)-1, k*(i+1)-1)], fill = Rec_color) 
del draw

display(img)
```

```python
days_count = 0
stats = [[days_count, n*n - 1, 1, 0]] # day, S, I, R

while (input() != '0'):
    temp_aut = [[Sus] * n for i in range(n)]
    
    stats.append(stats[days_count].copy())
    days_count += 1
    stats[days_count][0] += 1
    
    for i in range(n):
        for j in range(n):
            if (aut[i][j] == Sus):
                # counting infected neighbors
                nbr = 0
#                 if (aut[(i - 1) % n][(j - 1) % n] == Inf): nbr += 1
                if (aut[(i - 1) % n][j] == Inf): nbr += 1
#                 if (aut[(i - 1) % n][(j + 1) % n] == Inf): nbr += 1
                if (aut[i][(j - 1) % n] == Inf): nbr += 1
                if (aut[i][(j + 1) % n] == Inf): nbr += 1
#                 if (aut[(i + 1) % n][(j - 1) % n] == Inf): nbr += 1
                if (aut[(i + 1) % n][j] == Inf): nbr += 1
#                 if (aut[(i + 1) % n][(j + 1) % n] == Inf): nbr += 1
                    
                if (nbr == 1 and random.uniform(0,1) < 0.3\
                    or nbr == 2 and random.uniform(0,1) < 0.6\
                    or nbr > 2):
                    temp_aut[i][j] = Inf
                    inf_days[i][j] = Inf_period
                    
                    stats[-1][1] -= 1
                    stats[-1][2] += 1
                    
            elif (aut[i][j] == Inf):
                if (inf_days[i][j] != 0):
                    temp_aut[i][j] = Inf
                    inf_days[i][j] -= 1
                else:
                    temp_aut[i][j] = Rec
                    
                    stats[-1][2] -= 1
                    stats[-1][3] += 1
                    
            elif (aut[i][j] == Rec):
                temp_aut[i][j] = Rec
                    
    aut = temp_aut.copy()
    
    img = Image.new('RGB', (size, size), color = 'white')
    
    draw = ImageDraw.Draw(img)
    
    for i in range(n):
        for j in range(n):
            if (aut[i][j] == Sus): draw.rectangle([(k*j, k*i), (k*(j+1)-1, k*(i+1)-1)], fill = Sus_color) 
            elif (aut[i][j] == Inf): draw.rectangle([(k*j, k*i), (k*(j+1)-1, k*(i+1)-1)], fill = Inf_color)
            elif (aut[i][j] == Rec): draw.rectangle([(k*j, k*i), (k*(j+1)-1, k*(i+1)-1)], fill = Rec_color) 
    del draw
    
    display(img)
```

```python
stats_days = []
stats_s = []
stats_i = []
stats_r = []
for i in range(days_count):
    stats_days.append(stats[i][0])
    stats_s.append(stats[i][1])
    stats_i.append(stats[i][2])
    stats_r.append(stats[i][3])
```

```python
plt.plot(stats_days, stats_s)
plt.plot(stats_days, stats_i)
plt.plot(stats_days, stats_r)
plt.show()
```
