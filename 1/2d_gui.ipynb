```python
from PyQt5 import QtWidgets, QtGui
from design2 import Ui_MainWindow
import sys
import time

import random
from PIL import Image, ImageDraw
```

```python
n = 128
size = 512
k = size // n

flag = 0

aut = []
```

```python
class mywindow(QtWidgets.QMainWindow):
    def __init__(self):
        super(mywindow, self).__init__()
        self.ui = Ui_MainWindow()
        self.ui.setupUi(self)
        
        self.ui.pushButton.clicked.connect(self.btnClicked)
        self.ui.pushButton_2.clicked.connect(self.btnClicked2)
        self.ui.pushButton_3.clicked.connect(self.btnClicked3)
        self.ui.pushButton_4.clicked.connect(self.btnClicked4)
        self.ui.pushButton_5.clicked.connect(self.btnClicked5)
        self.ui.pushButton_6.clicked.connect(self.btnClicked6)
        self.ui.pushButton_7.clicked.connect(self.btnClicked7)
        
    def btnClicked(self):
        # random + preview
        global aut

        aut = [[0] * n for i in range(n)]
        for i in range(n):
            for j in range(n):
                if (random.randint(0, 1) == 0): aut[i][j] = 1
                    
        img = Image.new('RGB', (size, size), color = 'white')

        draw = ImageDraw.Draw(img)
        for i in range(n):
            for j in range(n):
                if (aut[i][j] == 1): draw.rectangle([(k*j, k*i), (k*(j+1)-1, k*(i+1)-1)], fill = "black") 
        del draw
        
        img.save('output.png')
        pix = QtGui.QPixmap('output.png')
        self.ui.label_3.setPixmap(pix)
        
    def btnClicked2(self):
        pass
        
    def btnClicked3(self):
        pass
        
    def btnClicked4(self):
        pass
        
    def btnClicked5(self):
        pass
        
    def btnClicked6(self):
        for i in range(100):
            self.btnClicked7()
            
    def btnClicked7(self):
        # simulate / resume + output
        global aut, flag
        flag = 0
        
        # start simulating
        
        temp_aut = [[0] * n for i in range(n)]

        for i in range(n):
            for j in range(n):
                # counting neighbors
                nbr = 0
                if (aut[(i - 1) % n][(j - 1) % n] == 1): nbr += 1
                if (aut[(i - 1) % n][j] == 1): nbr += 1
                if (aut[(i - 1) % n][(j + 1) % n] == 1): nbr += 1
                if (aut[i][(j - 1) % n] == 1): nbr += 1
                if (aut[i][(j + 1) % n] == 1): nbr += 1
                if (aut[(i + 1) % n][(j - 1) % n] == 1): nbr += 1
                if (aut[(i + 1) % n][j] == 1): nbr += 1
                if (aut[(i + 1) % n][(j + 1) % n] == 1): nbr += 1

                if (aut[i][j] == 0):
                    if (nbr == 3): temp_aut[i][j] = 1
                else:
                    if (nbr == 2 or nbr == 3): temp_aut[i][j] = 1

        aut = temp_aut.copy()

        img = Image.new('RGB', (size, size), color = 'white')

        draw = ImageDraw.Draw(img)
        for i in range(n):
            for j in range(n):
                if (aut[i][j] == 1): draw.rectangle([(k*j, k*i), (k*(j+1)-1, k*(i+1)-1)], fill = "black") 
        del draw

        #output overall picture

        img.save('output.png')
        pix = QtGui.QPixmap('output.png')
        self.ui.label_3.setPixmap(pix)

app = QtWidgets.QApplication([])
application = mywindow()
application.show()
 
sys.exit(app.exec())
```
