```python
from PyQt5 import QtWidgets, QtGui
from design import Ui_MainWindow
import sys

import random
from PIL import Image, ImageDraw
```

```python
def conv_rule(n):
    b = bin(n)
    return b.replace("0b","0" * int(10 - len(b)))
```

```python
n = 128
size = 512
k = size // n

aut = [0 for i in range(n)]
aut[0], aut[1], aut[2] = 1, 1, 1
# aut= []
```

```python
class mywindow(QtWidgets.QMainWindow):
    def __init__(self):
        super(mywindow, self).__init__()
        self.ui = Ui_MainWindow()
        self.ui.setupUi(self)
        
        self.ui.pushButton.clicked.connect(self.btnClicked)
        self.ui.pushButton_2.clicked.connect(self.btnClicked2)
        self.ui.pushButton_3.clicked.connect(self.btnClicked3)
        self.ui.pushButton_4.clicked.connect(self.btnClicked4)
        self.ui.pushButton_5.clicked.connect(self.btnClicked5)
        self.ui.pushButton_6.clicked.connect(self.btnClicked6)
        
    def btnClicked(self):
        # random
        global aut
        aut = [random.randint(0, 1) for i in range(n)]
        
    def btnClicked2(self):
        # all 1
        global aut
        aut = [1 for i in range(n)]
        
    def btnClicked3(self):
        # all 0
        global aut
        aut = [0 for i in range(n)]
        
    def btnClicked4(self):
        # one centered
        global aut
        aut = [0 for i in range(n)]
        aut[n // 2] = 1
        
    def btnClicked5(self):
        # one cornered
        global aut
        aut = [0 for i in range(n)]
        aut[n - 1] = 1
        
    def btnClicked6(self):
        # simulate + output
        global aut
        
        rule_n = int(self.ui.lineEdit.text())
        rule = conv_rule(rule_n)
        
        img = Image.new('RGB', (size, size), color = 'white')

        # output first line

        draw = ImageDraw.Draw(img)
        for i in range(n):
            if (aut[i] == 1): draw.rectangle([(k*i, 0), (k * (i + 1) - 1, k - 1)], fill = "black") 
        del draw
        
        # start simulating
        
        for i in range(n):
            temp_aut = [0] * n

            for j in range(n):
                if (aut[(j - 1) % n] == 1):
                    if (aut[j % n] == 1):
                        if (aut[(j + 1) % n] == 1):
                            temp_aut[j] = int(rule[0])
                        else:
                            temp_aut[j] = int(rule[1])
                    else:
                        if (aut[(j + 1) % n] == 1):
                            temp_aut[j] = int(rule[2])
                        else:
                            temp_aut[j] = int(rule[3])
                else:
                    if (aut[j % n] == 1):
                        if (aut[(j + 1) % n] == 1):
                            temp_aut[j] = int(rule[4])
                        else:
                            temp_aut[j] = int(rule[5])
                    else:
                        if (aut[(j + 1) % n] == 1):
                            temp_aut[j] = int(rule[6])
                        else:
                            temp_aut[j] = int(rule[7])

            aut = temp_aut.copy()

            # draw on picture

            draw = ImageDraw.Draw(img)
            for j in range(n):
                if (aut[j] == 1): draw.rectangle([(k*j, k*(i+1)), (k*(j+1)-1, k*(i+2)-1)], fill = "black") 
            del draw        
        
        #output overall picture
        
        img.save('output.png')
        pix = QtGui.QPixmap('output.png')
        self.ui.label_3.setPixmap(pix)

app = QtWidgets.QApplication([])
application = mywindow()
application.show()
 
sys.exit(app.exec())
```
